```python
from simulate import *
from animate import *
from IPython.display import HTML, display
import warnings
warnings.filterwarnings(action='ignore', module='strawberryfields')
```

# No interaction case

```python
F = np.array([[1, 0], [0, 1]])
masses = np.array([3, 1])
# F = np.eye(2)
K = 100
dt = .2
L = 6
xvec, pvec = np.linspace(-L, L, 500), 5*np.linspace(-L, L, 500)
eng = sf.Engine('gaussian')
x0 = [-3, -1]
simulator = quadratic_simulator(F, masses)

snapshots1, snapshots2 = [], []
progress = tqdm(total=K)
for k in range(K):
    res1, res2 = simulator.run(k, dt, xvec, pvec, eng, x0)
    snapshots1.append(res1)
    snapshots2.append(res2)
    progress.update(1)
progress.close()
```

```python
anim = animate_evolution(K, xvec, snapshots1, snapshots2, label1='Particle 1', label2='Particle 2', interval=80)

# anim.save('./animations/two_particle.gif', writer='ffmpeg')
video = HTML(anim.to_html5_video())
display(video)
plt.close()
```
